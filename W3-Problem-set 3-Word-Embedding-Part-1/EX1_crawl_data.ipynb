```python
import pandas as pd
```

```python
from datasets import load_dataset

ds = load_dataset("wikimedia/wikipedia", "20231101.en", streaming= True)
```

```python
ds = ds['train']
```

```python
ds = ds.take(10000)
```

```python
df = pd.DataFrame.from_dict(ds)
```

```python
df = df.sample(50)
```

```python
res = df
```

```python
res
```

```python
res.to_csv("W1/data/all_data.csv", index= False)
```

```python
for idx, row in res.iterrows():
    title = row['title']
    with open(f"W1/data/{title}.txt", "w") as f:
        f.write(row['text'])
```
